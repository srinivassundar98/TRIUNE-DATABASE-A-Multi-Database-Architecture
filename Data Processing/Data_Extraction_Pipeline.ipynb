```python
##MAINS
import json
import os
from datetime import datetime

# Placeholder sets for tracking seen tweets and users
seen_tweets = set()
seen_users = set()

output_directory = "corona/"  # Set the output directory to 'corona'
os.makedirs(output_directory, exist_ok=True)  # Create the directory if it doesn't exist

file_paths = ["corona-out-2/corona-out-2", "corona-out-3/corona-out-3"]  # List of file paths to process

def format_timestamp(date_str):
    return datetime.strptime(date_str, '%a %b %d %H:%M:%S +0000 %Y').strftime('%Y%m%d%H%M%S')

for file_path in file_paths:
    with open(file_path, "r", encoding='utf-8') as f1:
        for line in f1:
            line = line.strip()
            if not line:
                continue

            try:
                data = json.loads(line)
            except json.JSONDecodeError as e:
                print(f"Error decoding JSON for line: {line[:40]}... Error: {e}")
                continue

            tweet_id = data['id_str']
            timestamp = format_timestamp(data['created_at'])

            # Check if the tweet has been seen before
            if tweet_id in seen_tweets:
                continue
            seen_tweets.add(tweet_id)

            user = data['user']
            user_id = user['id_str']

            # Check if the user has been seen before
            if user_id not in seen_users:
                seen_users.add(user_id)
                # Add user to user table (or any other datastore you are using)

            # Determine the prefix based on whether it's a tweet or retweet
            prefix = 'RT' if data['text'].startswith('RT') else 'T'

            # Save the tweet or retweet as a JSON file
            file_name = f"{prefix}_{tweet_id}_{timestamp}.json"
            with open(os.path.join(output_directory, file_name), 'w', encoding='utf-8') as outfile:
                json.dump(data, outfile, ensure_ascii=False, indent=4)
```
